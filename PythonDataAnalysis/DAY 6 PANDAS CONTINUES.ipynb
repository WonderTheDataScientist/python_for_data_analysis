#### REMOVING OUTLIERS IN DATASET 
Outliers are unusual data points in your dataset


```python
houseData = {'lotsize':[2.0, 6.0, 7.0, 5.1, 8.0], 'roomsize':[12, 40.5, 60.8, 50.9, 900], 
             'Price':[2000, 6000, 7000, 5200, 90000]}

dfhou = pd.DataFrame(houseData)
dfhou
```

```python
# dfhou.scatter(x = lotsize, y=price)\
dfhou.plot(kind="bar")
```

```python
# import matplotlib.pyplot as plt
dfhou.plot.scatter(x = "lotsize", y = "Price")
```

```python
lower_threshold =  dfhou['lotsize'].quantile(0.05)
lower_threshold
```

```python
upper_threshold =  dfhou['lotsize'].quantile(0.95)
upper_threshold
```

```python
lotsmall = dfhou[dfhou['lotsize'] > lower_threshold]
lotsmall
```

```python
lotbig = dfhou[dfhou['lotsize'] < upper_threshold]
lotbig
```

```python
realdata = dfhou[(dfhou['lotsize'] > lower_threshold) & (dfhou['lotsize'] < upper_threshold)]
realdata
```

```python
realdata.plot.scatter(x = 'lotsize', y='Price')
```
